```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='chainName',data=df)
```

### 체인별 cap

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'cap', data=df, ax=ax1)
# chainName별 cap 최댓값
ax2 = sns.violinplot(x='chainName', y='cap', data=df, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'cap', data=df, estimator=np.max, ax=ax3)
# chainName별 cap 평균값
ax4 = sns.boxplot(x='chainName', y = 'cap', data=df, ax=ax4)

plt.show()
```

### 체인별 price

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'price', data=df, ax=ax1)
# chainName별 price 최댓값
ax2 = sns.violinplot(x='chainName', y='price', data=df, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'price', data=df, estimator=np.max, ax=ax3)
# chainName별 price 평균값
ax4 = sns.boxplot(x='chainName', y = 'price', data=df, ax=ax4)

plt.show()
```

### 체인별 Volume

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'totalVolume', data=df, ax=ax1)
# chainName별 totalVolume 최댓값
ax2 = sns.violinplot(x='chainName', y='totalVolume', data=df, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'totalVolume', data=df, estimator=np.max, ax=ax3)
# chainName별 totalVolume 평균값
ax4 = sns.boxplot(x='chainName', y = 'totalVolume', data=df, ax=ax4)

plt.show()
```

### 체인별 14일 단위 가격 변화율

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'priceChangePercentage14d', data=df, ax=ax1)
# chainName별 totalVolume 최댓값
ax2 = sns.violinplot(x='chainName', y='priceChangePercentage14d', data=df, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'priceChangePercentage14d', data=df, estimator=np.max, ax=ax3)
# chainName별 totalVolume 평균값
ax4 = sns.boxplot(x='chainName', y = 'priceChangePercentage14d', data=df, ax=ax4)

plt.show()
```

```python
cos_df = pd.DataFrame()
juno_df = pd.DataFrame()
kava_df = pd.DataFrame()
osmo_df = pd.DataFrame()
```

```python
cos_df = df[df['chainName']=='cosmos']
juno_df = df[df['chainName']=='juno']
kava_df = df[df['chainName']=='kava']
osmo_df = df[df['chainName']=='osmosis']

cos_df
```

## 확률밀도함수+히스토그램

### 코인별 cap 분포

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.distplot(cos_df['cap'],kde=True, hist=True, ax=ax1)
ax2 = sns.distplot(juno_df['cap'],kde=True, hist=True, ax=ax2)
ax3 = sns.distplot(kava_df['cap'],kde=True, hist=True, ax=ax3)
ax4 = sns.distplot(osmo_df['cap'],kde=True, hist=True, ax=ax4)

ax1.set_title('<Distplot of cap from cosmos>')
ax2.set_title('<Distplot of cap from juno>')
ax3.set_title('<Distplot of cap kava>')
ax4.set_title('<Distplot of cap osmosis>')

plt.show()
```

## 회귀선 + 산점도

### 코인별 cap과 Volume 상관관계

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.regplot(x='cap', y='totalVolume', data=cos_df, ax=ax1)
ax2 = sns.regplot(x='cap', y='totalVolume', data=juno_df, ax=ax2)
ax3 = sns.regplot(x='cap', y='totalVolume', data=kava_df, ax=ax3)
ax4 = sns.regplot(x='cap', y='totalVolume', data=osmo_df, ax=ax4)

ax1.set_title('<Scatterplot of cap&Volume from cosmos>')
ax2.set_title('<Scatterplot of cap&Volume from juno>')
ax3.set_title('<Scatterplot of cap&Volume from kava>')
ax4.set_title('<Scatterplot of cap&Volume from osmosis>')

plt.show()
```
