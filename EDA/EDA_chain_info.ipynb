```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='chainName',data=data)
```

### 체인별 apr

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'apr', data=data, ax=ax1)
# chainName별 apr 최댓값
ax2 = sns.violinplot(x='chainName', y='apr', data=data, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'apr', data=data, estimator=np.max, ax=ax3)
# chainName별 apr 평균값
ax4 = sns.boxplot(x='chainName', y = 'apr', data=data, ax=ax4)

plt.show()
```

### 체인별 communityTokens

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'communityTokens', data=data, ax=ax1)
# chainName별 communityTokens 최댓값
ax2 = sns.violinplot(x='chainName', y='communityTokens', data=data, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'communityTokens', data=data, estimator=np.max, ax=ax3)
# chainName별 communityTokens 평균값
ax4 = sns.boxplot(x='chainName', y = 'communityTokens', data=data, ax=ax4)

plt.show()
```

### 체인별 tps

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.pointplot(x='chainName', y = 'tps', data=data, ax=ax1)
# chainName별 tps 최댓값
ax2 = sns.violinplot(x='chainName', y='tps', data=data, ax=ax2)
ax3 = sns.pointplot(x='chainName', y = 'tps', data=data, estimator=np.max, ax=ax3)
# chainName별 tps 평균값
ax4 = sns.boxplot(x='chainName', y = 'tps', data=data, ax=ax4)

plt.show()
```

```python
cos_df = pd.DataFrame()
juno_df = pd.DataFrame()
kava_df = pd.DataFrame()
osmo_df = pd.DataFrame()
```

```python
cos_df = data[data['chainName']=='cosmos']
juno_df = data[data['chainName']=='juno']
kava_df = data[data['chainName']=='kava']
osmo_df = data[data['chainName']=='osmosis']

cos_df
```

## 확률밀도함수+히스토그램

### 토큰별 apr 분포

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.distplot(cos_df['apr'],kde=True, hist=True, ax=ax1)
ax2 = sns.distplot(juno_df['apr'],kde=True, hist=True, ax=ax2)
ax3 = sns.distplot(kava_df['apr'],kde=True, hist=True, ax=ax3)
ax4 = sns.distplot(osmo_df['apr'],kde=True, hist=True, ax=ax4)

ax1.set_title('<apr distplot of cosmo>')
ax2.set_title('<apr distplot of juno>')
ax3.set_title('<apr distplot of kava>')
ax4.set_title('<apr distplot of osmosis>')

plt.show()
```

### 토큰별 communityToken 분포

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.distplot(cos_df['communityTokens'],kde=True, hist=True, ax=ax1)
ax2 = sns.distplot(juno_df['communityTokens'],kde=True, hist=True, ax=ax2)
ax3 = sns.distplot(kava_df['communityTokens'],kde=True, hist=True, ax=ax3)
ax4 = sns.distplot(osmo_df['communityTokens'],kde=True, hist=True, ax=ax4)

ax1.set_title('<communityTokens distplot of cosmo>')
ax2.set_title('<communityTokens distplot of juno>')
ax3.set_title('<communityTokens distplot of kava>')
ax4.set_title('<communityTokens distplot of osmosis>')

plt.show()
```

### 토큰별 tps 분포

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
ax1, ax2, ax3, ax4 = axs.flatten()

ax1 = sns.distplot(cos_df['tps'],kde=True, hist=True, ax=ax1)
ax2 = sns.distplot(juno_df['tps'],kde=True, hist=True, ax=ax2)
ax3 = sns.distplot(kava_df['tps'],kde=True, hist=True, ax=ax3)
ax4 = sns.distplot(osmo_df['tps'],kde=True, hist=True, ax=ax4)

ax1.set_title('<tps distplot of cosmo>')
ax2.set_title('<tps distplot of juno>')
ax3.set_title('<tps distplot of kava>')
ax4.set_title('<tps distplot of osmosis>')

plt.show()
```

## 산점도

### 코인별 communitytoken 과 apr 산포도
